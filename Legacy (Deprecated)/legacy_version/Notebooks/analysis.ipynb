```python
import seaborn as sns
import pandas as pd
import matplotlib as mp
```

```python
sns.set_theme()
pitchdata = pd.read_csv('pitch_data.csv')
pitchdata['Pitcher'] = pitchdata['Pitcher'].map(lambda x: ''.join(ch for ch in x if ch.isalnum()))
pitchdata['PitchType'] = pitchdata['PitchType'].map(lambda x: ''.join(ch for ch in x if ch.isalnum()))
```

```python
plot = sns.relplot(data=pitchdata,
                   x='FinalX', y='FinalY',
                   hue='PitchType',
                   col='Pitcher',
                   col_wrap=2,
                   kind='scatter',
                   height=4, aspect=1.5)
for ax in plot.axes:
    ax.set_xlim(400, 900)
    ax.set_ylim(300, 700)
    ax.set(ylim=ax.get_ylim()[::-1])
    ax.add_patch(mp.patches.Rectangle((565, 410), 130,
                        150,fill=False,edgecolor='black',lw=1))
```

Plot by PitchTypes

```python
plot2 = sns.relplot(data=pitchdata,
                   x='FinalX', y='FinalY',
                   col='PitchType',
                   col_wrap=2,
                   kind='scatter',
                   height=4, aspect=1.5)
for ax in plot2.axes:
    ax.set_xlim(300, 950)
    ax.set_ylim(250, 750)
    ax.set(ylim=ax.get_ylim()[::-1])
    ax.add_patch(mp.patches.Rectangle((565, 410), 130,
                        150,fill=False,edgecolor='black',lw=1))
```

```python
salePitches = pitchdata[pitchdata['Pitcher'] == 'chrissale']
salePlot = sns.relplot(data=salePitches,
                   x='FinalX', y='FinalY',
                   col='PitchType',
                   hue='PitchType',
                   col_wrap=1,
                   kind='scatter',
                   height=4,
                   aspect=1.5,
                   s=110)
for ax in salePlot.axes:
    ax.set_xlim(300, 900)
    ax.set_ylim(300, 700)
    ax.set(ylim=ax.get_ylim()[::-1])
    ax.add_patch(mp.patches.Rectangle((565, 410), 130,
                        150,fill=False,edgecolor='black',lw=1))


```
